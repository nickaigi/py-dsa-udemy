## Write a program to add two numbers

```python
def solution(num1: int, num2: int) -> int:
    return num1 + num2
```

```python
import unittest

class TestExample(unittest.TestCase):
    def test(self) -> None:
        self.assertEqual(solution(2, 2), 4)
        print('All Tests passed')
```

```python
t = TestExample()
t.test()
```
